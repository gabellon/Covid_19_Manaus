### Importação e visualização de informações básicas dos dados  ####

```python
# import pandas
import pandas as pd

# leitura do dataset
df = pd.read_csv('./Resources/BlackFriday.csv')

#informações basicas da estrutura dos dados
display(df.info())

#preview dos dados
display(df.head(10))
```

### Classificação das Variáveis ####

```python
tipo_variaveis = [  ['User_ID', 'Quantitativa Discreta'],
                    ['Product_ID', 'Categórica Nominal'],
                    ['Gender', 'Categórica Nominal'],
                    ['Age', 'Categórica Ordinal'],
                    ['Occupation', 'Categórica Nominal'],
                    ['City_Category', 'Categórica Nominal'],
                    ['Stay_In_Current_City_Years', 'Categórica Ordinal'],
                    ['Marital_Status', 'Categórica Nominal'],
                    ['Product_Category_1', 'Categórica Nominal'],
                    ['Product_Category_2', 'Categórica Nominal'],
                    ['Product_Category_3', 'Categórica Nominal'],
                    ['Purchase', 'Quantitativa Discreta']]

tipo_variaveis = pd.DataFrame(tipo_variaveis, columns=['Variável', 'Classificação'])

display(tipo_variaveis)
```

## Tratamento de nulos

```python
# Preenchimento dos valores NaN com 0, sendo as colunas com NaN Product_Category_2 e Product_Category_3, e
# sendo elas categóricas e numéricas, então todos os valores NaN podem ser representados por 0
# para manter o tipo numérico em todas as linhas para que o python possa trabalhar sem erro de tipo de dados

df.fillna(value=0,inplace=True)

display(df.head(5))
```

### Análise Exploratória: Entendento o dataset

#### Percentual de compras por gênero

```python
import matplotlib.pyplot as plt

# titulos dos gêneros no gráfico 
labels = 'Masculino', 'Feminino'

# contagem dos totais de registros separados pelo gênero
sizes = [df['Gender'][df['Gender'] == 'M'].count(), df['Gender'][df['Gender'] == 'F'].count()]

# pequeno distanciamento entre as partes do gráfico
explode = (0, 0.1) 

# montagem do gráfico
fig1, ax1 = plt.subplots()

ax1.set_title('Percentual de compras por gênero')

ax1.pie(sizes,
        explode=explode,
        labels=labels, 
        autopct='%1.1f%%', #conversão automática em percentual
        shadow=True,
        startangle=90)

ax1.axis('equal')  

plt.show()
```

#### Percentual de compras por estado civil

```python
labels = 'Solteiros', 'Casados'
sizes = [df['Marital_Status'][df['Marital_Status'] == 0].count(), df['Marital_Status'][df['Marital_Status'] == 1].count()]
explode = (0, 0.1) 

fig1, ax1 = plt.subplots()

ax1.set_title('Percentual de compras por estado civil')

ax1.pie(sizes,
        explode=explode,
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)

ax1.axis('equal')  

plt.show()
```

#### Percentual de compras por faixa etária

```python
# crio um dataframe com a contagem dos valores separadamente
a1 = df['Age'].value_counts()

# ordeno e reseto o index para que o indice atual (faixas etárias) se torne uma coluna selecionável
a1 = a1.to_frame().sort_index().reset_index()

# pego cada eleento da coluna de faixa etária como label do grágico
labels = a1['index'].astype('str').to_list()

# pego cada contagem de cada faixa etaria correspondente
sizes = a1['Age'].to_list()

plt.figure(figsize=(10,10))

plt.pie(sizes,
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)

plt.title('Percentual de compras por faixa etária')

plt.show()
```

#### Percentual de compras por categoria de cidade

```python
c1 = df['City_Category'].value_counts()

c1 = c1.to_frame().sort_index().reset_index()

labels = c1['index'].astype('str').to_list()
sizes = c1['City_Category'].to_list()
explode = (0.05, 0.05,0.1) 

plt.pie(sizes,
        explode=explode,
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)

plt.title('Percentual de compras por categoria de cidade')

plt.show()
```

### Exemplo de Relação entre as Variáveis

#### Relação de compras entre a categoria das cidades e a Product_Category_1

O seguinte gráfico demonstra as compras em cada uma das três cidades amostradas e agrupadas por cada categoria da coluna Product_Category_1, quase todas as categorias e cidades apresentam o mesmo comportamento nas compras, onde A é o tipo de cidade que menos comprou, C a intermediária e B o tipo que mais comprou, um comportamento diretamente proporcional ao percentual de compras por cidade demonstrado no gráfico acima, exceto nas categorias "3", "12" e "7". Uma diferença que, pela grande quantidade dos dados e pela complexidade de análise, sem o auxílio de gráficos e do processamento computacional seria imperceptível. Diferença que pode representar uma oportunidade de negócio em relação às cidades e produtos em questão.

```python
import numpy as np

# agrupo o numero de compras por categoria e cidade e reseto o indice, pois a coluna de cidade se tornaria o indice
# tornando a cidade e o Product_Category_1 em colunas selecionaveis
# coluna de User_ID sendo utilizada somente como coluna unica para contagem 
gp = pd.DataFrame(df.groupby(['City_Category','Product_Category_1'])['User_ID'].count()).reset_index()

# agrupo o numero de compras por categoria e reseto o indice somente para pegar as categorias como labels
g1 = df['Product_Category_1'].value_counts().to_frame().reset_index()

labels = g1['index'].astype('str')
xl = np.arange(len(labels))
width = 0.35  # largura das barras

# coluna de User_ID sendo utilizada somente como coluna unica para contagem 
sizeA = gp[gp['City_Category'] == 'A']['User_ID'].to_list()
sizeB = gp[gp['City_Category'] == 'B']['User_ID'].to_list()
sizeC = gp[gp['City_Category'] == 'C']['User_ID'].to_list()

# montagem do gráfico
fig, ax = plt.subplots(figsize=(12,10))

# crio defasagem igual entre os 3 tipos de cidade para evitar sobreposição das barras
rects1 = ax.bar(xl-(width+0.2)/3, sizeA,width, label='Cidade A',color='r')
rects2 = ax.bar(xl,         sizeB,width, label='Cidade B',color='g')
rects3 = ax.bar(xl+(width+0.2)/3, sizeC,width, label='Cidade C',color='b')

# Titulos e legendas
ax.set_ylabel('Nº de compras')
ax.set_xlabel('Cidades por categoria')
ax.set_title('Relação de compras entre a categoria das cidades e a Product_Category_1')
ax.set_xticks(xl)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
